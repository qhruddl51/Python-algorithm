# 문제 ) n이하의 소수를 나열하기

```python
def prime (n : int) :
    prime_list = []
    for num in range(1,n+1) : 
        counter = 0
        
        for i in range(2,num) :
            if num % i == 0 :  # -> 소수가 아니다. 2 이상의 어떠한 제수로도 나누어 떨어지면 안된다.  
                counter += 1
            pass # inner for
        
        if counter == 0 :
            prime_list.append(num)
    
    return prime_list


prime(15)        
```

```python
counter = 0 

for n in range(2, 15) :
    
    for i in range(2,n) :
        counter += 1
        if n%i == 0 :
            print(f'n = {n}, i = {i}일 때, 반복문을 빠져나옵니다.')
            break
            
    else : 
        print(n)# else 위치 중요! if 조건에 한 번도 안 걸렸었다면, 이 시점에서 n을 출력한다.
        
print(f'나눗셈을 실행한 횟수: {counter}')
```

###  알고리즘 개선 1 - 홀수만을 대상으로 하기, 1부터 n-1까지의 모든 수가 아니라 기존에 구한 소수들로 나눠보기 

```python
counter = 0
ptr = 0
prime_list = [None]*500

prime_list[ptr]=2 #  2는 소수이므로 초깃값으로 지정
ptr += 1

for n in range(3, 1001, 2) : # 홀수만을 대상으로 설정
    
    for i in range(1, ptr) :
        counter += 1
        if n%prime_list[i] == 0 :
            break
    
    else : 
        prime_list[ptr] = n # 소수로 배열에 등록
        ptr += 1 
        

for i in range(ptr) : # 소수를 출력
    print(prime_list[i])
    
print(f'나눗셈을 실행한 횟수 : {counter}')
```

###  알고리즘 개선 2 - 소수가 제곱근보다 작은 경우까지만 연산할 수 있도록 하기

```python
counter = 0
ptr = 0
prime_list = [None]*500

prime_list[ptr] = 2 # 0번째 소수는 2이다. 
ptr += 1

prime_list[ptr] = 3 # 1번째 소수는 3이다. 
ptr += 1

for n in range(5,1001,2) :
    i = 0 # i를 다시 초기화 
    while prime_list[i]*prime_list[i] <= n :
        counter += 2 # 곲셈횟수 : prime_list[i]*prime_list[i], n % prime_list[i] 
        
        if n % prime_list[i] == 0 :
            break
        
        i+=1
        
    else : # while 반복문을 다 돌면서 한번도 if 조건에 걸리지 않았었다면, prime_list에 ptr번째 소수로 n을 추가
        prime_list[ptr] = n
        ptr += 1
        counter +=1 # 곲셈횟수 : prime_list[i]*prime_list[i]
        
        
for i in range(ptr) :
    print(prime_list[i])
    
print(f'곱셈과 나눗셈을 실행한 횟수 : {counter}')
```
